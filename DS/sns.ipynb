```python
import seaborn as sns
df= sns.load_dataset('flights')
df.head()
```

```python
# line chart 
sns.lineplot(data= df,x='year',y='passengers')
```

```python
a=sns.lineplot(data= df,x='year',y='passengers', hue='month')
a.set_title('Passenger Count in different Years')
a.set_xlabel('Year')
a.set_ylabel('Passengers Count')
# a.set_xticklabels()
```

```python
titanic= sns.load_dataset('titanic')
titanic.head()
```

```python
#count plot 
sns.countplot(data=titanic, x='pclass')
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(data=titanic, x='pclass', y='age', hue='sex')
```
