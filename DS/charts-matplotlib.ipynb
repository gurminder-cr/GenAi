```python
# barplot 
# one side numerical value other side categorical value 
x= ['Anurag','Bhawan','Anil','Ansh','Nikhil']
marks=[34,39,43,29,33]

plt.bar(x=x, height=marks)
plt.show()
```

```python
import pandas as pd 
df= pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python
df.Pclass.value_counts()
```

```python
plt.bar(df.Pclass.value_counts().index, df.Pclass.value_counts().values)
plt.xlabel('Pclass')
plt.ylabel('Count of Passengers')
plt.title('Count of Passengers for every passenger class')
plt.show()
```

```python
plt.barh(df.Pclass.value_counts().index, df.Pclass.value_counts().values)
plt.title('Count of Passengers for every passenger class')
plt.show()
```

```python
plt.barh(df.Pclass.value_counts().index, df.Pclass.value_counts().values, color='pink')
```

```python
# Scatter Plot 
# continuous- numerical values of both sides
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x,y)
```
